```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import joblib
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
train_df = pd.read_csv('train.csv')
dev_df = pd.read_csv('val.csv')
test_df = pd.read_csv('test.csv')

train_df
     
```

```python
X_train = train_df.drop(columns=["label"])
Y_train = train_df["label"]

X_dev = dev_df.drop(columns=["label"])
Y_dev = dev_df["label"]

X_test = test_df.drop(columns=["label"])
Y_test = test_df["label"]
```

```python
import mlflow
from train_with_mlflow import log_experiment_run  
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier


mlflow.set_tracking_uri("http://127.0.0.1:5000")
mlflow.set_experiment("Hand_Gesture_Classification")

gesture_labels=['call', 'dislike', 'fist', 'four', 'like', 'mute', 'ok', 'one',
       'palm', 'peace', 'peace_inverted', 'rock', 'stop', 'stop_inverted',
       'three', 'three2', 'two_up', 'two_up_inverted']
```

```python
runs = []

# ------ SVM ------
kernels = ['linear', 'rbf', 'poly']
c_values = [0.1, 1, 10, 50, 100]
degrees = [2, 3, 4]
gammas = ['scale', 0.1, 1]

for k in kernels:
    c_values = [100] if k == 'poly' else c_values
    for c in c_values:
        current_gammas = gammas if k in ['rbf', 'poly'] else ['scale']
        for g in current_gammas:
            current_degrees = degrees if k == 'poly' else [3]
            for d in current_degrees:
              runs.append({ "name": f"SVM_{k}","class": SVC, "params": {"C": c, "kernel": k, "gamma":g, "degree":d} })

```

```python
# --- RANDOM FOREST  ---
for n in [50,100, 200]:
    for d in [ None,10, 20,30,50]:
        runs.append({
            "name": f"RF__n_{n}_D_{d}",
            "class": RandomForestClassifier,
            "params": {"n_estimators": n, "max_depth": d, "random_state": 42} })
```

```python
from sklearn.neighbors import KNeighborsClassifier

# ---- KNN ----
for n in [3, 5, 7, 11, 21]:
        runs.append({ "name": f"KNN__n_{n}", "class": KNeighborsClassifier, "params": {"n_neighbors": n} })
```

```python
for run in runs:
    print(f"Training: {run['name']}")
    
    ##initialize the model and training it
    model = run["class"](**run["params"])
    model.fit(X_train, Y_train)
    
    #Log to MLflow
    m = log_experiment_run( model, X_train, Y_train, X_dev, Y_dev, run_name=run["name"], params=run["params"], gesture_labels=gesture_labels )
    
print("Done logging")
```
